```python
#removing the columns
df=df.drop('BuildingArea',axis=1)
df=df.drop('YearBuilt',axis=1)
df=df.drop('CouncilArea',axis=1)

round(100*(df.isnull().sum()/len(df.index)),2)
```

```python
#retaining the rows haveing 5 or less null values as 18% deletion wont affect the results
df=df[df.isnull().sum(axis=1)<=5]

round(100*(df.isnull().sum()/len(df.index)),2)
```

```python
#20% of price column is sitll missing and it is the target column i.e I am not going to use it but will use other attributes to predict the prize.
df=df[~np.isnan(df.Price)]

round(100*(df.isnull().sum()/len(df.index)),2)
```

```python
df["Landsize"].describe()
# the min is 0,max is 433333, the mean is 558 and median 440. the data looks skewed there's too much variation in 25th and 75 percentile as well
#thus imputing with with mean/median seems quite biased, so lets remove them
```

```python
df=df[~np.isnan(df.Landsize)]
round(100*(df.isnull().sum()/len(df.index)),2)
```

```python
df.isnull().sum()
```

```python
df[["Lattitude","Longtitude"]].describe()
```

```python
df[np.isnan(df["Lattitude"])]=df.Lattitude.mean()
df.isnull().sum()
```

```python
#the distributuion of lat and long is quite narrow. A good way to estimate way to estimate the spread of data is to look at the difference b/w the mean and median(lower the better) and variation from 25 to 75 percentile(quite small in this case)
#thus I am imputing the missing values by the mean value of lat and long

df.loc[np.isnan(df["Lattitude"]),['Lattitude']]=df.Lattitude.mean()
df.loc[np.isnan(df["Longtitude"]),["Longtitude"]]=df.Longtitude.mean()
round(100*(df.isnull().sum()/len(df.index)),2)
```
